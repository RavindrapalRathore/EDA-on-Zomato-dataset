```python
ratings=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

```python
ratings.head()
```

**OBSERVATION**



1.   When Rating is between from 4.5 to 4.9 the rating were excelent
2.   When Rating is between from 4.0 to 4.4 the rating were very good
2.   When Rating is between from 3.5 to 3.9 the rating were good
2.   When Rating is between from 2.5 to 3.4 the rating were average
2.   When Rating is between from 1.8 to 2.4 the rating were poor
2.   When Rating is 0 it means people has not given any rating

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(x='Aggregate rating', y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','green','yellow','pink','orange'])
```

**OBSERVATION**



1.   Not rated count is very high
2.   Maximum number of rating are between 2.5 to 3.4



```python
#count Plot

sns.countplot(x="Rating color", data=ratings)
```
